```python
def ACF_and_PACF(data):
  fig = plt.figure(figsize=(17,10))
  ax1 = fig.add_subplot(211)
  fig = plot_acf(data , lags = 30 , ax = ax1)
  ax2 = fig.add_subplot(212)
  fig = plot_pacf(data , lags = 30 , ax = ax2)

ACF_and_PACF(shows_added)
```

### It seems that AC is slowly and PAC is quickly so we may need AR model


# Find order d value through differencing


```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize':(9,7), 'figure.dpi':120})


# Original Series
fig, axes = plt.subplots(3, 2, sharex=True)
axes[0, 0].plot(shows_added); axes[0, 0].set_title('Original Series')
plot_acf(shows_added, ax=axes[0, 1])

# 1st Differencing
axes[1, 0].plot(shows_added.diff()); axes[1, 0].set_title('1st Order Differencing')
plot_acf(shows_added.diff().dropna(), ax=axes[1, 1])

# 2nd Differencing
axes[2, 0].plot(shows_added.diff().diff()); axes[2, 0].set_title('2nd Order Differencing')
plot_acf(shows_added.diff().diff().dropna(), ax=axes[2, 1])

plt.show()
```

#### Our data stationary in th first differencing so d=1


```python
decomposition = sm.tsa.seasonal_decompose(shows_added , model = "additive").plot()
plt.plot()
```

```python
#Determine rolling statistics
plt.figure(figsize=(15,7))

rolmean = shows_added.rolling(window=12).mean() #window size 12 denotes 12 months, giving rolling mean at yearly level
rolstd = shows_added.rolling(window=12).std()
#Plot rolling statistics
orig = plt.plot(shows_added, color='blue', label='Original')
mean = plt.plot(rolmean, color='red', label='Rolling Mean')
std = plt.plot(rolstd, color='black', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

# Find order p value through PAC


```python
# PACF plot of 1st differenced series
plt.rcParams.update({'figure.figsize':(9,3), 'figure.dpi':120})

plt.plot(shows_added.diff());
plt.xlabel('1st Differencing')
plot_pacf(shows_added.diff().dropna())

plt.show()
```

#### We can see that the PACF lag 1 is quite significant since it is well above the significance line. So, we will fix the value of p as 1


# Find order q value through AC


```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plt.plot(shows_added.diff());
plt.xlabel('1st Differencing')
plot_acf(shows_added.diff().dropna())

plt.show()
```

#### We can see that one of lags is well above the significance line so q is 1


# Auto Arima Forecasting in Python

```python
from statsmodels.tsa.arima_model import ARIMA

import pmdarima as pm

model = pm.auto_arima(shows_added, start_p=1, start_q=1,
                      test='adf',       # use adftest to find optimal 'd'
                      max_p=3, max_q=3, # maximum p and q
                      m=1,              # frequency of series
                      d=None,           # let model determine 'd'
                      seasonal=True,   # No Seasonality
                      start_P=0,
                      D=0,
                      trace=True,
                      error_action='ignore',
                      suppress_warnings=True,
                      stepwise=True)

print(model.summary())
```

#### The p-values are highly significant (<< 0.05)


# Builing our model


```python
shows_added.shape
```

```python
from statsmodels.tsa.stattools import acf

# Create Training and Test
train = shows_added[:63]
test = shows_added[63:]
```

```python
# Build Model
model = sm.tsa.arima.ARIMA(train, order=(1,1,1))
fitted = model.fit()

# Forecast
fc= fitted.forecast(steps=30, alpha=0.05)  # 95% conf

# Make as pandas series
fc_series = pd.Series(fc)

# Plot
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train, label='training')
plt.plot(test, label='actual')
plt.plot(fc_series, label='forecast')
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```
